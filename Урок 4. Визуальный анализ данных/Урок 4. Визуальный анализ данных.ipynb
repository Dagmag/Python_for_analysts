```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('laptops_10.csv')
df.head()
```

> Анализ распределения компаний. Количество ноутбуков в разрезе компаний.

```python
plt.figure(figsize=(10,5))
plt.hist(df['Company'], bins=35, color='r')
plt.title('Количество ноутбуков в разрезе компаний', fontsize=14)
plt.xlabel('Company')
plt.ylabel('Counts')
plt.grid()
plt.xticks(rotation = 45);
```

> По графику мы види 3 основные лидирующие компании Dell, Lenovo и HP.

```python
plt.figure(figsize=(10,5))
plt.hist(df['TypeName'], bins=10, color='m')
plt.title('Типы ноутбуков', fontsize=16)
plt.xlabel('Type', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.xticks(rotation = 45)
plt.grid(axis='y',color='k', linestyle='-', linewidth=0.5);
```

> Вывод: Самое большое количество простых "Ноутбуков", далее идет "Ноутбук для игр" и потом уже "Ультрабук" (Лёгкий ноутбук)

```python
plt.figure(figsize=(10,5))
plt.hist(df['Price_euros'],bins=250, color='c')
plt.title('Количество ноутбуков по цене', fontsize=16)
plt.xlabel('Price', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.grid(axis='y');
```

> Основное количество ноутбуков находится в ценовом диапазоне до 1500 ЕВРО

```python
date = df['Memory_Type'].value_counts()
date
```

```python
names = date.index
values = date.values
explode = (0.05, 0.05, 0.15, 0.3)
plt.pie(values, explode=explode, labels = names, autopct ='%.0f%%', startangle=0)

plt.title('Накопители', fontsize=16)
plt.legend(loc='upper left');
```
