```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('laptops_10.csv')
df.head()
```

> Анализ распределения компаний. Количество ноутбуков в разрезе компаний.

```python
plt.figure(figsize=(10,5))
plt.hist(df['Company'], bins=35, color='r', edgecolor = 'black')
plt.title('Количество ноутбуков в разрезе компаний', fontsize=14)
plt.xlabel('Company')
plt.ylabel('Counts')
plt.grid(axis='x')
plt.xticks(rotation = 45);
```

> По графику мы види 3 основные лидирующие компании Dell, Lenovo и HP.

```python
plt.figure(figsize=(10,5))
plt.hist(df['TypeName'], bins=10, color='m')
plt.title('Типы ноутбуков', fontsize=16)
plt.xlabel('Type', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.xticks(rotation = 45)
plt.grid(axis='y',color='k', linestyle='-', linewidth=0.5);
```

> Вывод: Самое большое количество простых "Ноутбуков", далее идет "Ноутбук для игр" и потом уже "Ультрабук" (Лёгкий ноутбук)

```python
plt.figure(figsize=(10,5))
plt.hist(df['Price_euros'],bins=250, color='c')
plt.title('Анализ цен', fontsize=16)
plt.xlabel('Price', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.grid(axis='y');
```

> Основное количество ноутбуков находится в ценовом диапазоне до 1500 ЕВРО

```python
date = df['Memory_Type'].value_counts()
date
```

```python
names = date.index
values = date.values
explode = (0.05, 0.05, 0.15, 0.3)
plt.pie(values, explode=explode, labels = names, autopct ='%.0f%%', startangle=0)

plt.title('Накопители', fontsize=16)
plt.legend(loc='upper left');
```

> У большенства Ноутбуках установлена память SSD

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Company'], df['Inches'], c='g',s=50, marker='+')
plt.scatter(df['Company'], df['Weight'], c='r',s=10)
plt.title('Соотношение размера ноутбука к его весу, по бренду', fontsize=16)
plt.xticks(rotation = 45)
plt.grid(axis='x');
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

- id - Уникальный ID для каждого дома
- date - Дата продажи дома
- price - Стоимость продажи дома
- bedrooms - Кол-во спален
- bathrooms - Кол-во ванных комнат (0.5 - туалет без душа)
- sqft_living - Кв. метры жилые
- sqft_lot - Кв. метры общие
- floors - Кол-во этажей
- waterfront - Есть набержная или нет
- view - Значение от 0 до 4 насколько хороший вид
- condition - Значение от 1 до 5 насколько хорошее состояние
- grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
- sqft_above - Кв. метры дома, которые находятся выше земли
- sqft_basement - Кв. метры дома, которые находятся ниже земли
- yr_built - Год постройки дома
- yr_renovated - Год ремонта дома
- zipcode - Индекс
- lat - Широтма
- long - Долгота
- sqft_living15 - Кв. метры жилой площади у 15 соседей
- sqft_lot15 - Кв. метры общей площади у 15 соседей

#### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(10,5))
plt.hist(df['price'], color='b', bins=200)
plt.title('Стоимость недвижимости', fontsize=16)
plt.ylabel('Цена', fontsize=14)
plt.grid(axis='x');
```

#### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(10,5))
plt.hist(df['sqft_living'], color='mediumseagreen', bins=200, edgecolor = 'black')
plt.title('Количество этажей по домам', fontsize=16)
plt.ylabel('Кв. метры жилые', fontsize=14);
```

#### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10,5))
plt.hist(df['yr_built'], color='tab:olive', bins=100, edgecolor = 'black')
plt.title('Посторойка домов', fontsize=16)
plt.xlabel('год', fontsize=14)
plt.ylabel('Количество', fontsize=14)
plt.grid(axis='y');
```

## Условие 2: 2 задача

- 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
data = df['waterfront'].value_counts()
data
```

```python
names = data.index
values = data.values
explode = (0.5, 0.05)
plt.pie(values, explode=explode, labels = names, autopct ='%.0f%%', startangle=30)

plt.title('Вид на набережную', fontsize=16)
plt.legend(loc='upper left');
```

```python
data_view = df['view'].value_counts()
data_view
```

```python
names = data_view.index
values = data_view.values
explode = (0.3 , 0, 0, 0, 0)
plt.pie(values, labels = names, autopct ='%.0f%%', startangle=30, explode = explode, shadow = True)

plt.title('На сколько хороший вид из квартиры', fontsize=16)
plt.legend(loc='upper left');
```
